### 1. How many customers exist per city, and what is their average account balance?

```python
query_1 = """SELECT city, COUNT(*) AS customers, 
AVG(account_balance) AS avg_balance
FROM customers c
JOIN accounts a USING (customer_id)
GROUP BY city LIMIT 5;
"""

result = pd.read_sql_query(query_1, conn)

print(result)
```

### 2. Which account type (Savings, Current, Loan, etc.) holds the highest total balance? 

```python
query_2 = """SELECT 
    c.account_type,
    SUM(a.account_balance) AS total_balance
FROM customers c
JOIN accounts a 
    ON c.customer_id = a.customer_id
GROUP BY c.account_type
ORDER BY total_balance DESC;"""

result = pd.read_sql_query(query_2, conn)

print(result)
```

### 3. Who are the top 10 customers by total account balance across all account types?

```python
query_3 = """SELECT 
    c.customer_id,
    c.name,
    c.city,
    c.account_type,
    SUM(a.account_balance) AS total_balance
FROM customers c
JOIN accounts a 
    ON c.customer_id = a.customer_id
GROUP BY c.customer_id, c.name, c.city, c.account_type
ORDER BY total_balance DESC
LIMIT 5;"""

result = pd.read_sql_query(query_3, conn)

print(result)
```

### 4. Which customers opened accounts in 2023 with a balance above ₹1,00,000

```python
query_4 = """SELECT 
    c.customer_id,
    c.name,
    c.city,
    c.account_type,
    a.account_balance,
    c.join_date
FROM customers c
JOIN accounts a 
    ON c.customer_id = a.customer_id
WHERE strftime('%Y', c.join_date) = '2023'
  AND a.account_balance > 100000
ORDER BY a.account_balance DESC LIMIT 5 ;"""

result = pd.read_sql_query(query_4, conn)

print(result)
```

### 5. What is the total transaction volume (sum of amounts) by transaction type

```python
query_5 = """SELECT 
    txn_type,
    SUM(amount) AS total_transaction_volume
FROM transactions
GROUP BY txn_type
ORDER BY total_transaction_volume DESC;
"""

result = pd.read_sql_query(query_5, conn)

print(result)
```

### 6. Which accounts have more than 3 failed transactions in a single month?

```python
query_6 = """SELECT 
    customer_id,
    strftime('%Y-%m', txn_time) AS txn_month,
    COUNT(*) AS failed_txn_count
FROM transactions
WHERE status = 'failed'
GROUP BY customer_id, txn_month
HAVING COUNT(*) > 2
ORDER BY failed_txn_count DESC LIMIT 5;
"""

result = pd.read_sql_query(query_6, conn)

print(result)
```

### 7. Which are the top 5 branches by total transaction volume in the last 6 months?

```python
query_7 = """SELECT
    c.name,
    B.Branch_Name,
    SUM(T.amount) AS total_transaction_volume
FROM
Transactions T
JOIN Customers C ON T.customer_id = C.customer_id
JOIN Branches B ON C.city = B.City  
WHERE
    T.txn_time >= strftime('%Y-%m-%d %H:%M:%S', 'now', '-6 months')
GROUP BY B.Branch_Name
ORDER BY total_transaction_volume DESC
LIMIT 5;

"""

result = pd.read_sql(query_7, conn)

print(result)
```

###  8. Which accounts have 5 or more high-value transactions above ₹9999

```python
query_8 = """SELECT
    T.customer_id,
    COUNT(T.txn_id) AS high_value_transaction_count
FROM
    Transactions T
WHERE
    T.amount >= 95000 
    AND T.status = 'success' 
GROUP BY
    T.customer_id
HAVING
    COUNT(T.txn_id) >= 5
ORDER BY
    high_value_transaction_count DESC LIMIT 5;"""
result = pd.read_sql_query(query_8, conn)

print(result)
```

### 9. What is the average loan amount and interest rate by loan type (Personal, Auto, Home, etc.)?

```python
query_9 = """SELECT 
    Loan_Type,
    AVG(Loan_Amount) AS avg_loan_amount,
    AVG(Interest_Rate) AS avg_interest_rate
FROM loans
GROUP BY Loan_Type
ORDER BY avg_loan_amount DESC;
"""

result = pd.read_sql_query(query_9, conn)

print(result)
```

### 10. Which customers currently hold more than one active or approved loan?

```python
query_10 = """SELECT
    Customer_ID,
    COUNT(*) AS active_loans
FROM loans
WHERE Loan_Status IN ('Active', 'Approved')
GROUP BY Customer_ID
HAVING COUNT(*) > 1
ORDER BY active_loans DESC LIMIT 5;
"""

result = pd.read_sql_query(query_10, conn)

print(result)
```

### 11. Who are the top 5 customers with the highest outstanding (non-closed) loan amounts?

```python
query_11 = """SELECT
    Customer_ID,
    SUM(Loan_Amount) AS total_outstanding
FROM loans
WHERE Loan_Status != 'Closed'
GROUP BY Customer_ID
ORDER BY total_outstanding DESC
LIMIT 5;
"""

result = pd.read_sql_query(query_11, conn)

print(result)
```

### 12. Which branch holds the highest total account balance?

```python
query_12 = """SELECT 
    b.Branch_Name,
    SUM(a.account_balance) AS total_balance
FROM accounts a
JOIN customers c 
    ON a.customer_id = c.customer_id
JOIN branches b
    ON c.city = b.City
GROUP BY b.Branch_Name
ORDER BY total_balance DESC
LIMIT 5;
"""

result = pd.read_sql_query(query_12, conn)

print(result)
```

### 13. What is the branch performance summary showing total customers, total loans, and transaction volume?

```python
query_13 = """SELECT 
    b.Branch_Name,
    COUNT(DISTINCT c.customer_id) AS total_customers,
    COUNT(DISTINCT l.Loan_ID) AS total_loans,
    SUM(t.amount) AS total_transaction_volume
FROM branches b
LEFT JOIN customers c ON c.city = b.City
LEFT JOIN loans l ON l.Branch = b.Branch_Name
LEFT JOIN transactions t ON t.customer_id = b.Branch_Name
GROUP BY b.Branch_Name;"""

result = pd.read_sql_query(query_13, conn)

print(result)
```

### 14. Which issue categories have the longest average resolution time?

```python
query_14 = """
SELECT 
    issue_category,
    AVG(JULIANDAY(date_closed) - JULIANDAY(date_opened)) AS avg_resolution_days
FROM tickets
WHERE date_opened IS NOT NULL
GROUP BY issue_category
ORDER BY avg_resolution_days DESC
"""

result = pd.read_sql_query(query_14, conn)

print(result)
```

### 15. Which support agents have resolved the most critical tickets with high customer ratings (≥4)?

```python
query_15 = """SELECT 
    support_agent,
    COUNT(*) AS resolved_critical_high_rated_tickets
FROM tickets
WHERE 
    priority = 'Critical'
    AND status = 'Resolved'
    AND customer_rating >= 4
GROUP BY support_agent
ORDER BY resolved_critical_high_rated_tickets DESC;
"""

result = pd.read_sql_query(query_15, conn)

print(result)
```
